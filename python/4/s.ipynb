```python
import random
def ranpass(l):
    char = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*" 
    password = ""
    for _ in range(l):
        password += random.choice(char)
    return password
print (ranpass(12))
```

```python
import random
res = {
    "hello" : ['hello','hi'],
    'how are you' : ["i'm well,doing fine"],
    "goodbye" : ["goodbye","see you later","farewell"],
    "default" : ["i didn't understand","could you please rephrase that"]
}

def getres(userinput):
    for key in res:
        if key in userinput:
            return random.choice(res[key])
    return random.choice(res["default"])

def chatbot():
    print('chatbot: how can i help you')
    while True:
        userinput=input("user:  ").lower()
        response = getres(userinput)
        print("chatbot: ", response)

        if userinput == 'goodbye':
            break
chatbot()
```
